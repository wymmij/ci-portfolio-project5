# Load and inspect data

The Kaggle archive extracts the CSV files inside a timestamped folder. I have kept the extracted raw data structure unchanged in `inputs/datasets/raw` to preserve the original source structure.

```python
from pathlib import Path
import pandas as pd

raw_data_dir = Path("inputs/datasets/raw")

house_prices_path = next(raw_data_dir.rglob("house_prices_records.csv"))
inherited_houses_path = next(raw_data_dir.rglob("inherited_houses.csv"))

print(house_prices_path)
print(inherited_houses_path)

house_prices_df = pd.read_csv(house_prices_path)
inherited_houses_df = pd.read_csv(inherited_houses_path)
```

Both duplicate checks returned zero, so no duplicated rows were found in either dataset.

For later notebooks, stable copies of the two project datasets are saved in `outputs/datasets/collection`.

```python
collection_dir = Path("outputs/datasets/collection")
collection_dir.mkdir(parents=True, exist_ok=True)

house_prices_df.to_csv(collection_dir / "house_prices_records.csv", index=False)
inherited_houses_df.to_csv(collection_dir / "inherited_houses.csv", index=False)
```
